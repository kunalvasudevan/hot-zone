```
grouped_test=D1[['symbol', 'average_price']].groupby(['symbol'])
grouped_test.head(2)
```

```
f_val, p_val = stats.f_oneway(grouped_test.get_group('AAPL')['average_price'], grouped_test.get_group('FANG')['average_price'], grouped_test.get_group('BBY')['average_price'],grouped_test.get_group('GE')['average_price'],grouped_test.get_group('COG')['average_price'],grouped_test.get_group('PVH')['average_price'],grouped_test.get_group('CDNS')['average_price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val)
```
